## Vectors

ðŸ“Œ What is a Vector?

A vector is a mathematical quantity that has:

Magnitude (size/length)

Direction

Example:
If you say "Walk 5 km towards the east,"

5 km = magnitude

east = direction

So, a vector tells us how much and which way.

Vectors are usually written like â†’AB or in coordinates as (x, y) in 2D or (x, y, z) in 3D.

ðŸ“Œ Types of Vectors

Vectors can be classified in many ways. Here are the important types:

1. Zero Vector (Null Vector)

Magnitude = 0

Written as (0,0) in 2D or (0,0,0) in 3D

No direction.
Example: Staying at the same place.

2. Unit Vector

Magnitude = 1

Shows only direction.
Example: (1, 0) is a unit vector along the x-axis.

Usually written with a hat: Ã®, Äµ, kÌ‚

3. Position Vector

Shows the position of a point from the origin.
 Example: Point A(3,4) has position vector â†’OA = (3,4).

4. Equal Vectors

Two vectors are equal if they have the same magnitude and direction, even if they start at different points.

5. Negative Vector

Opposite direction but same magnitude.
Example: â†’AB = (2,3) â†’ Negative vector = (âˆ’2, âˆ’3).

6. Collinear Vectors

Lie on the same line or parallel lines.
Example: (2,4) and (1,2) are collinear.

7. Like and Unlike Vectors

Like vectors â†’ same direction.

Unlike vectors â†’ opposite direction.

8. Co-planar Vectors

Vectors that lie in the same plane (2D or 3D plane).

9. Orthogonal (Perpendicular) Vectors

Two vectors are orthogonal if their dot product = 0.
Example: (1,0) and (0,1).

10. Displacement Vector

Represents movement from one point to another.
Example: If you move from (2,3) to (5,7), displacement vector = (3,4).

```python
# Define a 2D vector
vector = np.array([2, 3])

# Create a figure and a set of subplots
plt.figure(figsize=(8,4))
fig, ax = plt.subplots()

# Plot the vector as an arrow
ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1)

# Set limits and labels
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a grid
ax.grid()

# Show the plot
plt.show()
```

```python
# Define two 2D vectors
vector1 = np.array([2, 3])
vector2 = np.array([1, -2])

# Add the vectors
resultant_vector = vector1 + vector2

# Create a figure and a set of subplots
plt.figure(figsize=(8,4))
fig, ax = plt.subplots()

# Plot the vectors as arrows
ax.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
ax.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')
ax.quiver(0, 0, resultant_vector[0], resultant_vector[1], angles='xy', scale_units='xy', scale=1, color='g', label='Resultant Vector/ after addition')

# Set limits and labels
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Vector Addition')

# Add a grid and legend
ax.grid()
ax.legend()

# Show the plot
plt.show()
```

```python
# Define a 2D vector
vector = np.array([-2, 1])

# Define a scalar
scalar = 2

# Perform scalar multiplication
scaled_vector = vector * scalar

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot the vectors as arrows
ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
ax.quiver(0, 0, scaled_vector[0], scaled_vector[1], angles='xy', scale_units='xy', scale=1, color='b', label='Scaled Vector/After multiplication')

# Set limits and labels
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scalar Multiplication of a Vector')

# Add a grid and legend
ax.grid()
ax.legend()

# Show the plot
plt.show()
```

```python
# Define two 2D vectors
vector1_2d = np.array([2, 3])
vector2_2d = np.array([1, -2])

# Plot the 2D vectors for dot product context
fig_2d, ax_2d = plt.subplots()
ax_2d.quiver(0, 0, vector1_2d[0], vector1_2d[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
ax_2d.quiver(0, 0, vector2_2d[0], vector2_2d[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')
ax_2d.set_xlim([-5, 5])
ax_2d.set_ylim([-5, 5])
ax_2d.set_xlabel('X-axis')
ax_2d.set_ylabel('Y-axis')
ax_2d.set_title('Vectors for Dot Product')
ax_2d.grid()
ax_2d.legend()
plt.show()


# For 3D vectors, the cross product is a vector perpendicular to both original vectors.
vector3d_1 = np.array([1, 2, 3])
vector3d_2 = np.array([4, 5, 6])
cross_product_3d = np.cross(vector3d_1, vector3d_2)

# Plot 3D vectors and their cross product
fig_3d = plt.figure()
ax_3d = fig_3d.add_subplot(111, projection='3d')

ax_3d.quiver(0, 0, 0, vector3d_1[0], vector3d_1[1], vector3d_1[2], color='r', label='Vector 1')
ax_3d.quiver(0, 0, 0, vector3d_2[0], vector3d_2[1], vector3d_2[2], color='b', label='Vector 2')
ax_3d.quiver(0, 0, 0, cross_product_3d[0], cross_product_3d[1], cross_product_3d[2], color='g', label='Cross Product')


ax_3d.set_xlim([-10, 10])
ax_3d.set_ylim([-10, 10])
ax_3d.set_zlim([-10, 10])
ax_3d.set_xlabel('X-axis')
ax_3d.set_ylabel('Y-axis')
ax_3d.set_zlabel('Z-axis')
ax_3d.set_title('3D Vectors and their Cross Product')
ax_3d.legend()

plt.show()
```
